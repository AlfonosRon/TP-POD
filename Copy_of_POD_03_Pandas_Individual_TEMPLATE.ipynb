# Cancelaciones y Delays de vuelos del 2015

Creese una cuenta en Kaggle e importe los archivos del dataset del siguiente link: https://www.kaggle.com/datasets/usdot/flight-delays. Cree los dataframes `airlines`, `airports`, y `flights` apartir de esos archivos.

```python
#Escribir el codigo para responder la pregunta en esta celda
import pandas as pd

airlines= pd.read_csv("/content/airlines.csv")
airports= pd.read_csv("/content/airports.csv")
flights= pd.read_csv("/content/flights.csv")

aerolineas= pd.DataFrame(airlines)
aeropuertos= pd.DataFrame(airports)
vuelos= pd.DataFrame(flights)

```

Combine (*join*) las tablas `airlines`, `airports`, y `flights` en una sola tabla.

```python
#Escribir el codigo para responder la pregunta en esta ce
aerolineas.rename(columns={'IATA_CODE': 'IATA_CODE_aerolineas'}, inplace=True)
aerolineas.rename(columns={'AIRLINE': 'AIRLINE_aerolineas'}, inplace=True)
df1= vuelos.join(aeropuertos, how="left")
df2=df1.join(aerolineas, how="left")
df2
```

¿Cuantos vuelos fueron al aeropuerto JFK?

```python
#Escribir el codigo para responder la pregunta en esta celda

conteo_jfk = (df2['DESTINATION_AIRPORT'] == 'JFK').sum()
conteo_jfk
```

```python
#Escribir el codigo para responder la pregunta en esta celda
conteo_aa = (df2['AIRLINE'] == 'AA').sum()
conteo_aa
```

¿Que aerolineas (las primeras 5) tuvo la menor cantidad de vuelos con atrasos? Imprimirlas.

```python
#Escribir el codigo para responder la pregunta en esta celda
df2.dropna(subset=['DEPARTURE_DELAY', 'AIRLINE'], inplace=True)
UA = ((df2['DEPARTURE_DELAY'] > 0) & (df2['AIRLINE'] == 'UA')).sum()

AS = ((df2['DEPARTURE_DELAY'] > 0) & (df2['AIRLINE'] == 'AS')).sum()

AA = ((df2['DEPARTURE_DELAY'] > 0) & (df2['AIRLINE'] == 'AA')).sum()

US = ((df2['DEPARTURE_DELAY'] > 0) & (df2['AIRLINE'] == 'US')).sum()

F9 = ((df2['DEPARTURE_DELAY'] > 0) & (df2['AIRLINE'] == 'F9')).sum()

DL = ((df2['DEPARTURE_DELAY'] > 0) & (df2['AIRLINE'] == 'DL')).sum()

conteos={"UA":UA,"AS":AS,"AA":AA,"US":US,"F9":F9,"DL":DL}

aerolinea_minima = min(conteos, key=conteos.get)
aerolinea_minima
```
